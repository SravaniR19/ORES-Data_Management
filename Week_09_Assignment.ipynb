```python
import numpy as np
```

```python
Cancerrates_from_file['Male_incidence_counts'] = Cancerrates_from_file['Male_incidence_counts'].str.replace(',', '').astype(float)
Cancerrates_from_file.info()
```

```python
Cancerrates_from_file['Female_incident_counts'] = Cancerrates_from_file['Female_incident_counts'].str.replace(',', '').astype(float)
Cancerrates_from_file.info()
```

```python
Cancerrates_from_file
```

```python
Cancerrates_from_file.describe(include="all")
```

```python
Sum = ['Male_incidence_counts', 'Female_incident_counts']
Cancerrates_from_file['Total_sum'] = Cancerrates_from_file[Sum].sum(axis = 1)
```

```python
Cancerrates_from_file.head()
```

```python
Cancerrates_from_file.rename(columns = {'Female_incident_counts' : 'Female_incidence_counts', 'Total_sum' : 'Total_population_counts'}, inplace  = True)
```

```python
Cancerrates_from_file.head()
```

```python
from scipy import stats
z = np.abs(stats.zscore(Cancerrates_from_file['Male_incidence_counts']))
z = np.abs(stats.zscore(Cancerrates_from_file['Female_incidence_counts']))
threshold = 3
outliers = Cancerrates_from_file[z > threshold]
print(outliers)
```

```python
Cancerrates_from_file.duplicated()
```

```python
Cancerrates_from_file = Cancerrates_from_file.drop(outliers.index)
```
